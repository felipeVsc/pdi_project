```python
# teste de label missing vendo a quantidade de vermelho

# label missing

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./bottles/train/train_61.jpg')
#(thresh, im_bw) = cv.threshold(img, 128, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)
def bgr_to_rgb(image_bgr):
    # Use a funÃ§Ã£o cv2.cvtColor para converter de BGR para RGB
    image_rgb = cv.cvtColor(image_bgr, cv.COLOR_BGR2RGB)
    return image_rgb

img = bgr_to_rgb(img)

def contar_pixels_vermelhos(image_rgb):
    # Defina o intervalo de cores vermelhas (em RGB)
    lower_red = np.array([55, 0, 0])   
    upper_red = np.array([255, 0, 0]) 

    # Crie uma mÃ¡scara para os pixels vermelhos
    mask = cv.inRange(image_rgb, lower_red, upper_red)

    # Conte os pixels vermelhos na mÃ¡scara
    red_pixel_count = np.count_nonzero(mask)
    
    total_pixels = image_rgb.shape[0] * image_rgb.shape[1]

    # Calcule a porcentagem de pixels vermelhos
    percentage_red = (red_pixel_count / total_pixels) * 100

    return percentage_red

def calcular_porcentagem_vermelho_e_branco(imagem):
    # Converte a imagem para o espaÃ§o de cores HSV (Hue, Saturation, Value)
    hsv = cv.cvtColor(imagem, cv.COLOR_BGR2HSV)
    
    # Define os intervalos de cor para o vermelho e o branco
    # Note que esses intervalos podem variar dependendo da iluminaÃ§Ã£o e das condiÃ§Ãµes da imagem
    vermelho_min = np.array([70, 0, 0], dtype=np.uint8)
    vermelho_max = np.array([255, 0, 0], dtype=np.uint8)
    branco_min = np.array([150, 150, 150], dtype=np.uint8)
    branco_max = np.array([255, 255, 255], dtype=np.uint8)
    
    # Crie mÃ¡scaras para as cores vermelho e branco
    mascara_vermelha = cv.inRange(imagem, vermelho_min, vermelho_max)
    mascara_branca = cv.inRange(imagem, branco_min, branco_max)
    
    
    # Combina as mÃ¡scaras
    mascara_combinada = cv.bitwise_or(mascara_vermelha, mascara_branca)
    
    # Calcule o nÃºmero total de pixels na imagem
    total_pixels = imagem.shape[0] * imagem.shape[1]
    
    # Calcule o nÃºmero de pixels vermelhos e brancos
    pixels_vermelhos_e_brancos = np.count_nonzero(mascara_combinada)
    
    # Calcule a porcentagem de pixels vermelhos e brancos em relaÃ§Ã£o ao total
    porcentagem = (pixels_vermelhos_e_brancos / total_pixels) * 100.0
    
    return porcentagem

#count = contar_pixels_vermelhos(img)
count = calcular_porcentagem_vermelho_e_branco(img)
print(f"A quantidade de pixels vermelhos na imagem Ã©: {count}")

new_img = img[150:,100:260]
plt.imshow(new_img)
# label errada vai ser quando isso aqui tiver mt baixo
# bom fazer testes para ver como irÃ¡ se portar dado o crop feito

# Tem um limite superior e inferior, superior Ã© qnt nao tem mas tem liquido e inferior Ã© quando nao tem a botttle

# DOESNT WORK IN COLORS
```

```python
image = cv.imread('./bottles/train/train_11.jpg')
image = image[160:260,100:260]

# Convert the image to grayscale
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise and improve line detection
# Threshold the grayscale image to isolate white regions
_, thresholded = cv.threshold(gray, 150, 255, cv.THRESH_BINARY)

# Detect lines using the Hough Line Transform with theta set to np.pi/2
lines = cv.HoughLines(thresholded, 1, np.pi / 2, threshold=100)
print(lines)

# Draw the detected horizontal lines on the original image
if lines is not None:
    for line in lines:
        rho, theta = line[0]
        if np.pi / 4 <= theta <= 3 * np.pi / 4:  # Check if the line is vertical (theta is close to pi/2)
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a * rho
            y0 = b * rho
            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            cv.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

plt.imshow(image,'gray')
```
